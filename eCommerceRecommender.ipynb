# Simple Similarity Score Algorithm

```python
def getSimilarity(shoeRowNum1,shoeRowNum2):
    shoe1 = shoeData.loc[shoeRowNum1]
    shoe2 = shoeData.loc[shoeRowNum2]
    
    similarityScore = 0
    
    if shoe1['brand'] == shoe2['brand']:
        similarityScore += 1
    if shoe1['colors'] == shoe2['colors']:
        similarityScore += 1
    if shoe1['categories'] == shoe2['categories']:
        similarityScore += 1
    if shoe1['manufacturer'] == shoe2['manufacturer']:
        similarityScore += 1
        
    differenceInPrice = abs(float(shoe1['prices.amountMin']) - float(shoe2['prices.amountMin']))
                  
    if differenceInPrice <= 5:
        similarityScore += 3
    if differenceInPrice <= 25: 
        similarityScore += 2
    if differenceInPrice <= 45: 
        similarityScore += 1
    if differenceInPrice >= 45: 
        similarityScore += 0

    print(similarityScore)
```

# Compare a few shoes

```python
#TWO VERY DIFFERENT SHOES
getSimilarity(1004,2004)
```

```python
#TWO MORE SIMILAR SHOES
getSimilarity(218,2004)
```

```python
def getSimilaritiesForAll(shoeRowNum):
    numberOfShoes = len(shoeData)
    similarityScores = []
    
    
    for row in range(0,numberOfShoes):
        similarityScores.append(getSimilarity(shoeRowNum,row))
    
    return similarityScores
```
